audio signal parameters

```python
obj = wave.open('./Sample voice.wav', 'rb')
print(obj.getnchannels())
print(obj.getsampwidth())       # bytes per sample
print(obj.getframerate())
print(obj.getnframes())
print(obj.getparams())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import wave
```

```python
obj = wave.open('./Sample voice.wav', 'rb')
n_samples = obj.getnframes()
fr = obj.getframerate()

sample_wave = obj.readframes(-1)

time = n_samples/fr
sample_arr = np.frombuffer(sample_wave, dtype=np.int16)

t_axis = np.linspace(0, time, num=n_samples)

plt.figure(figsize=(15, 5))
plt.plot(t_axis, sample_arr)
```
